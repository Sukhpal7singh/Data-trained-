```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Country']=le.fit_transform(df['Country'])
```

Splitting and Training Data

```python
x = df.drop(['Happiness Score'],axis=1)
y = df['Happiness Score']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression

ln=LinearRegression()
ln.fit(x_train,y_train)
```

```python
y_pred=ln.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

print('Mean Absolute Error of linear regression:',mean_absolute_error(y_pred, y_test))
print('Mean Square Error of linear regression:', mean_squared_error(y_pred, y_test))
print('R_Squared Score of linear regression:', r2_score(y_pred, y_test))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

accuracy=ln.score(x_test,y_test)
print("Accuracy:",round(accuracy*100,2),"%")
```

```python
from sklearn import ensemble
rfr = ensemble.RandomForestRegressor()
rfr.fit(x_train,y_train)
pred = rfr.predict(x_test)
```

```python
rfr_accuracy=rfr.score(x_test,y_test)
print("Accuracy: ",round(rfr_accuracy*100,2),"%")
print('Mean Absolute Error of linear regression:',mean_absolute_error(pred, y_test))
print('Mean Square Error of linear regression:', mean_squared_error(pred, y_test))
print('R_Squared Score of linear regression:', r2_score(pred, y_test))
```
